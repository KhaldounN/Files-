```python
import pandas as pd
data = pd.read_csv('dataexport_20200921T130310.csv')
data
```

```python
date1,date2,date3,date4,date5 = [],[],[],[],[]

#data -> Raining & Snow & Basel Temperature [2 m elevation corrected] mean & cloud_cover_low
dato1, dato2, dato3, dato4, dato5 = [],[],[],[],[]

for i,d in enumerate((data["location"])):
    #print(data["location"][i], " : ",end="")
    #print(data["Basel.9"][i], "====",end="")
    #print(data["Basel.14"][i])
    #print(data["Basel.2"]
    
    if d.startswith("2015"):
        date1.append(d[:-5]) # to remove T0000 things [:-5]
        dato1.append([float(data["Basel.9"][i]),float(data["Basel.10"][i]),float(data["Basel.2"][i]),float(data["Basel.14"][i])])
       
    elif d.startswith("2016"):
        date2.append(d[:-5])
        dato2.append([float(data["Basel.9"][i]),float(data["Basel.10"][i]),float(data["Basel.2"][i]),float(data["Basel.14"][i])])
       
      
    elif d.startswith("2017"):
        date3.append(d[:-5])
        dato3.append([float(data["Basel.9"][i]),float(data["Basel.10"][i]),float(data["Basel.2"][i]),float(data["Basel.14"][i])])
       
     
    elif d.startswith("2018"):
        date4.append(d[:-5])
        dato4.append([float(data["Basel.9"][i]),float(data["Basel.10"][i]),float(data["Basel.2"][i]),float(data["Basel.14"][i])])
       
      
    elif d.startswith("2019"):
        date5.append(d[:-5])
        dato5.append([float(data["Basel.9"][i]),float(data["Basel.10"][i]),float(data["Basel.2"][i]),float(data["Basel.14"][i])])
       
```
